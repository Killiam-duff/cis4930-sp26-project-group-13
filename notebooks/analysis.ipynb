# Cleaning and Transformation

```python
#missing values
df.isnull().sum()
```

```python
#duplicate rows
print(f"Duplicate rows: {df.duplicated().sum()}")

#drop duplicates
df = df.drop_duplicates()
print(f"Shape after dropping duplicates: {df.shape}")
```

```python
#cleaning player names
df["Player"] = df["Player"].str.replace(r"[#*]", "", regex=True)

#someone was named Adeiny HechavarrÃƒÂ­a (handle weird characters in player names)
df["Player"] = df["Player"].str.encode("latin1", errors="ignore").str.decode("utf-8", errors="ignore")

df["Player"].sample(10)
```

```python
#make sure the batting average is numeric
df["BA"] = pd.to_numeric(df["BA"], errors="coerce")

#fill NaN values in BA with the median
df["BA"] = df["BA"].fillna(df["BA"].median())

#check for missing player names and drop those rows
df = df.dropna(subset=["Player"])

print(f"Shape after cleaning: {df.shape}")
df.dtypes
```

```python
#above average and below average players based on OPS
league_avg_ops = df["OPS"].mean()
df["OPS_tier"] = np.where(df["OPS"] >= league_avg_ops, "Above Avg", "Below Avg")

df[["Player", "Team", "OPS", "OPS_tier"]].head(10)
```

```python
#multi-column aggregation
#average stats by league
grouped = df.groupby("Lg").agg(
    avg_BA=("BA", "mean"),
    avg_HR=("HR", "mean"),
    avg_OPS=("OPS", "mean"),
    player_count=("Player", "count")
)
grouped
```

```python
#pivot table
#average HR by league and OPS tier
pivot = df.pivot_table(index="Lg", columns="OPS_tier", values="HR", aggfunc="mean")
pivot
```

```python
#export cleaned data
df.to_csv("../data/processed/batting_stats_cleaned.csv", index=False)
print("Cleaned data saved to data/processed/batting_stats_cleaned.csv")
```

# Visualizations

```python
import os
import matplotlib.ticker as mticker

os.makedirs("figures", exist_ok=True)

# shared style constants
sns.set_theme(style="whitegrid", palette="muted")
ACCENT  = "#1f77b4"
ACCENT2 = "#ff7f0e"
PALETTE = {"Above Avg": "#2ca02c", "Below Avg": "#d62728"}
```

## Figure 1: Distribution of OPS
Histogram with KDE overlay showing how OPS is distributed across all player-seasons, with mean and median reference lines.

```python
# Figure 1: Distribution of OPS (histogram + KDE)
fig, ax = plt.subplots(figsize=(9, 5))

sns.histplot(df_clean["OPS"], bins=40, kde=True, color=ACCENT,
             edgecolor="white", linewidth=0.4, ax=ax)

mean_ops = df_clean["OPS"].mean()
med_ops  = df_clean["OPS"].median()
ax.axvline(mean_ops, color="red",    linestyle="--", linewidth=1.5,
           label=f"Mean  {mean_ops:.3f}")
ax.axvline(med_ops,  color="orange", linestyle=":",  linewidth=1.5,
           label=f"Median {med_ops:.3f}")

ax.set_title("Distribution of OPS Across All Player-Seasons", fontsize=14, fontweight="bold")
ax.set_xlabel("OPS (On-base Plus Slugging)", fontsize=11)
ax.set_ylabel("Count", fontsize=11)
ax.legend(fontsize=10)
plt.tight_layout()
plt.savefig("../figures/fig1_ops_distribution.png", dpi=300)
plt.show()
```

## Figure 2: Avg Stolen Bases & OPS by Age
Dual-axis line chart illustrating how both stolen bases and OPS trend across player ages (restricted to ages with ≥10 player-seasons for reliability).

```python
# Figure 2: Avg SB & Avg OPS by Age (dual-axis line chart)
age_grp = df_clean.groupby("Age")[["SB", "OPS"]].mean().reset_index()
age_counts = df_clean.groupby("Age")["Player"].count()
valid_ages = age_counts[age_counts >= 10].index
age_grp = age_grp[age_grp["Age"].isin(valid_ages)]

fig, ax1 = plt.subplots(figsize=(10, 5))
ax2 = ax1.twinx()

ln1 = ax1.plot(age_grp["Age"], age_grp["SB"],  color=ACCENT,  marker="o",
               linewidth=2, label="Avg Stolen Bases")
ln2 = ax2.plot(age_grp["Age"], age_grp["OPS"], color=ACCENT2, marker="s",
               linewidth=2, linestyle="--", label="Avg OPS")

ax1.set_xlabel("Age", fontsize=11)
ax1.set_ylabel("Avg Stolen Bases", fontsize=11, color=ACCENT)
ax2.set_ylabel("Avg OPS",          fontsize=11, color=ACCENT2)
ax1.tick_params(axis="y", labelcolor=ACCENT)
ax2.tick_params(axis="y", labelcolor=ACCENT2)

lns  = ln1 + ln2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc="upper right", fontsize=10)
ax1.set_title("Avg Stolen Bases & OPS by Player Age\n(ages with ≥10 player-seasons)",
              fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("../figures/fig2_age_sb_ops.png", dpi=300)
plt.show()
```

## Figure 3: Home Runs vs WAR by OPS Tier
Scatter plot showing the relationship between home run totals and Wins Above Replacement, coloured by whether a player was above or below the league-average OPS.

```python
# Figure 4: Scatter HR vs WAR, coloured by OPS tier
PALETTE = {"Above Avg": "#2ca02c", "Below Avg": "#d62728"}

fig, ax = plt.subplots(figsize=(9, 6))

for tier, grp in df_clean.groupby("OPS_tier"):
    ax.scatter(grp["HR"], grp["WAR"], label=tier,
               color=PALETTE[tier], alpha=0.45, s=18, edgecolors="none")

ax.set_xlabel("Home Runs (HR)", fontsize=11)
ax.set_ylabel("Wins Above Replacement (WAR)", fontsize=11)
ax.set_title("Home Runs vs WAR  –  coloured by OPS Tier", fontsize=13, fontweight="bold")
ax.legend(title="OPS Tier", fontsize=10)
plt.tight_layout()
plt.savefig("../figures/fig4_hr_vs_war_ops_tier.png", dpi=300)
plt.show()
```

## Figure 4: AL vs NL League Comparison
Side-by-side bar charts comparing average Batting Average, OPS, and Home Runs between the American League and National League.

```python
# Figure 5: AL vs NL league comparison bar chart (avg BA, OPS, HR)
import matplotlib.ticker as mticker

league_grp = (df_clean[df_clean["Lg"].isin(["AL", "NL"])]
              .groupby("Lg")[["BA", "OPS", "HR"]]
              .mean()
              .reset_index())

metrics = ["BA", "OPS", "HR"]
labels  = ["Batting Average", "OPS", "Home Runs"]
fig, axes = plt.subplots(1, 3, figsize=(11, 5))

for ax, metric, label in zip(axes, metrics, labels):
    bars = ax.bar(league_grp["Lg"], league_grp[metric],
                  color=[ACCENT, ACCENT2], width=0.5, edgecolor="white")
    ax.bar_label(bars, fmt="%.3f" if metric in ("BA", "OPS") else "%.1f",
                 padding=3, fontsize=10)
    ax.set_title(label, fontsize=12, fontweight="bold")
    ax.set_xlabel("League", fontsize=10)
    ax.set_ylim(0, league_grp[metric].max() * 1.18)
    ax.yaxis.set_major_formatter(
        mticker.FormatStrFormatter("%.3f") if metric in ("BA", "OPS") else mticker.FormatStrFormatter("%.1f")
    )

fig.suptitle("AL vs NL – Average Batting Statistics", fontsize=14, fontweight="bold", y=1.02)
plt.tight_layout()
plt.savefig("../figures/fig5_league_comparison.png", dpi=300, bbox_inches="tight")
plt.show()
```

## Figure 5: OPS Distribution by Tier
Box plot comparing the spread of OPS values for above-average and below-average players, with the overall league mean shown as a dashed reference line.

```python
# Figure 6: Box-plot OPS distribution by OPS tier
PALETTE = {"Above Avg": "#2ca02c", "Below Avg": "#d62728"}

fig, ax = plt.subplots(figsize=(7, 5))
order = ["Above Avg", "Below Avg"]
sns.boxplot(data=df_clean, x="OPS_tier", y="OPS", order=order,
            hue="OPS_tier", palette=PALETTE, legend=False,
            width=0.45, linewidth=1.2,
            flierprops=dict(marker="o", markersize=3, alpha=0.3),
            ax=ax)
ax.axhline(df_clean["OPS"].mean(), color="black", linestyle="--",
           linewidth=1.2, label=f"League mean ({df_clean['OPS'].mean():.3f})")
ax.set_title("OPS Distribution by Tier", fontsize=13, fontweight="bold")
ax.set_xlabel("OPS Tier", fontsize=11)
ax.set_ylabel("OPS",      fontsize=11)
ax.legend(fontsize=10)
plt.tight_layout()
plt.savefig("../figures/fig6_ops_boxplot_by_tier.png", dpi=300)
plt.show()
```
